```python
nums = np.arange(2, 10001, dtype="int")
 
# sym.primefactors returns sorted list so the largest primenumber is always last
maxes = [ sym.primefactors(val)[-1] for val in nums ]
 
# Get dictionary with frequency counts: {prime1: frequency1, prime2: frequency2, ...}
freq = Counter(maxes)
 
# Create pandas dataframe for easy use of cumsum later
df = pd.DataFrame.from_dict(freq, orient="index", columns=["frequency"])
 
# Graph settings
_, ax = plt.subplots(figsize=(8,6))
ax.plot(df.index, df.frequency.cumsum(), "k.")
ax.axis((0, 500, 0, 10000))
ax.set_xlabel("Smooth number")
ax.set_ylabel("Number of occurences")
ax.set_title("K-smooth number frequencies")
plt.show()
```
