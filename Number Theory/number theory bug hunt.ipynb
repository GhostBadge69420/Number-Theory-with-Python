```python
ints = np.arange(1,101)

z = np.zeros((len(ints),len(ints)))

#run the simulations
for x in ints:  
    for y in ints:
        z[x-1,y-1] = (x**1 + y**2)**(1/3)

#check for integers
zInts = z%1 == 0

#now visualize
fig,ax = plt.subplots(1,2,figsize=(12,6))
fig.tight_layout(pad=5)

#show the resulting z-values
h = ax[0].imshow(z)
fig.colorbar(h,ax=ax[0],fraction=.045)
ax[0].set_xlabel('y')
ax[0].set_ylabel('x')
ax[0].set_title('$z=\sqrt[3]{x^1+y^2}$',fontsize=16)

#show the boolean integer map
h = ax[1].imshow(zInts,vmin=0,vmax=1)
fig.colorbar(h,ax=ax[1],fraction=.045)
ax[1].set_xlabel('y')
ax[1].set_ylabel('x')
ax[1].set_title('Whether $z$ is int',fontsize=1)

plt.show()
```

```python
a = np.arange(1,21)

s1 = np.zeros(len(a))
s2 = np.zeros(len(a))

for a in a:
      s1[a-1] = (-1)**a / np.log(np.exp(-a))
      s2[a-1] = np.sqrt(a) / ((-np.pi/2)**a)

fig,ax = plt.subplots(1,2,figsize=(10,4))
ax[0].plot(s1)
ax[0].set_title('Sequence $s_1$')
ax[1].plot(s2)
ax[1].set_title('Sequence $s_2$')

plt.show()
```
